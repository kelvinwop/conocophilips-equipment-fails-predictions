```python
import pandas as pd

df_training_full = pd.read_csv("training.csv", na_values=['na'])

df_training_full.head()
```

```python
df_test = pd.read_csv("test.csv", na_values=['na'])
df_test.head()
```

```python
#Combined both Train and Test Data set to do preprocessing together # and set flag for both as well
df_training_full['Type'] = 'Train' 
df_test['Type'] = 'Test'
fullData = pd.concat([df_training_full,df_test],axis=0,sort=True)
fullData.dtypes
```

```python
fullData.head()
```

```python
numeric_cols = list()
for k,v in fullData.iteritems():
    if v.dtype in ['int64', 'float64']:
        numeric_cols.append(k)


fullData[numeric_cols] = fullData[numeric_cols].fillna(fullData[numeric_cols].mean())
fullData.head()
```

```python
print("numeric columns:", numeric_cols)
# this spams
```

```python
# normalize
fullData[numeric_cols] = fullData[numeric_cols]/fullData[numeric_cols].max()
fullData.head()
```

```python
from sklearn.model_selection import train_test_split

train=fullData[fullData['Type']=="Train"]
test=fullData[fullData['Type']=="Test"]

numeric_cols.remove("target") #no answers kek
X = train[numeric_cols].values
y = train["target"].values

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
# CREATE HIGH LEVEL OVERVIEW METHODS
import random
from sklearn.ensemble import RandomForestRegressor
from tqdm import tqdm_notebook
from joblib import Parallel, delayed
import keras
import keras.layers
import numpy as np

def pmethod(xt, yt, est=10, subset=True, verbose=0):
    if subset:
        x_subset, _, y_subset, _ = train_test_split(xt, yt, test_size=0.1)
    else:
        x_subset, y_subset = xt, yt
        
    print("training with %s rows with %s targets" % (len(xt),sum(yt)))
    print("training with subset %s rows with %s targets" % (len(x_subset),sum(y_subset)))
    model = keras.Sequential()
    dropout = 0.5
    model.add(keras.layers.Dense(150, input_dim=np.shape(xt)[1], activation=keras.layers.LeakyReLU(alpha=0.2)))
    model.add(keras.layers.Dropout(dropout))
    model.add(keras.layers.Dense(1000, activation= keras.layers.LeakyReLU(alpha=0.4)))
    model.add(keras.layers.Dropout(dropout))
    model.add(keras.layers.Dense(100, activation= keras.layers.LeakyReLU(alpha=0.4)))
    model.add(keras.layers.Dropout(dropout))
    model.add(keras.layers.Dense(16, activation= keras.layers.LeakyReLU(alpha=0.4)))
    model.add(keras.layers.Dropout(dropout))
    model.add(keras.layers.Dense(16, activation= keras.layers.LeakyReLU(alpha=0.4)))
    model.add(keras.layers.Dropout(dropout))
    model.add(keras.layers.Dense(16, activation= keras.layers.LeakyReLU(alpha=0.4)))
    model.add(keras.layers.Dropout(dropout))
    model.add(keras.layers.Dense(8, activation= keras.layers.LeakyReLU(alpha=0.2)))
    model.add(keras.layers.Dense(4, activation= keras.layers.LeakyReLU(alpha=0.2)))
    model.add(keras.layers.Dense(1, activation='sigmoid'))
    class_weight={0:1, 1:(len(yt)-sum(yt))//sum(yt)//8}
    model.compile(loss= "binary_crossentropy" , optimizer="adam", metrics=["accuracy"])
    # rf = RandomForestRegressor(n_estimators=est)
    model.fit(x_subset, y_subset, epochs=10, verbose=verbose, class_weight=class_weight)
    # rf.fit(x_subset, y_subset)
    return model
    
def pforest(xt, yt, subset=True, est=10):
    if subset:
        x_subset, _, y_subset, _ = train_test_split(xt, yt, test_size=0.7)
    else:
        x_subset, y_subset = xt, yt
    rf = RandomForestRegressor(n_estimators=est)
    rf.fit(x_subset, y_subset)
    return rf

from sklearn.metrics import accuracy_score

def rn(n, cutoff=0.8):
    if n > cutoff:
        return 1
    else:
        return 0

def calc_stats(Vx, Vy, cutoff=0.8):
    """
    Vx: predictions
    Vy: actual
    """
    assert len(Vx) == len(Vy), "length mismatch: %s/%s" % (len(Vx), len(Vy))
    cp=0;cn=0;fp=0;fn=0
    for i in range(len(Vx)):
        if rn(Vx[i], cutoff=cutoff) == 1 and rn(Vy[i], cutoff=cutoff) == 1:
            cp+=1
        elif rn(Vx[i], cutoff=cutoff) == 0 and rn(Vy[i], cutoff=cutoff) == 0:
            cn+=1
        elif rn(Vx[i], cutoff=cutoff) == 1 and rn(Vy[i], cutoff=cutoff) == 0:
            fp+=1
        elif rn(Vx[i], cutoff=cutoff) == 0 and rn(Vy[i], cutoff=cutoff) == 1:
            fn+=1
        else:
            raise Exception("wtf")
    return cp, cn, fp, fn

def coag_superforest(sf, data, cutoff=0.8):
    # print("data length: ", len(data))
    predict_train_list = list()
    for rf in rf_list:
        predict_train_list.append(rf.predict(data))

    predict_train = list()
    for i in range(len(data)):
        s=0
        for pt in predict_train_list:
            s+=pt[i]
        predict_train.append(rn(s/len(predict_train_list), cutoff=cutoff))
    return predict_train

def get_superforest_predictions(rf_list, dataset, solutions=None, cutoff=0.8, get_acc=False, do_print=True, acc_met="as"):
    predict_train = coag_superforest(rf_list, dataset, cutoff=cutoff)
    if solutions is None:
        return predict_train
    tp, tn, fp, fn = calc_stats(predict_train, solutions, cutoff=cutoff)
    if do_print:
        print('Targets found on data %.2f/%.2f (total:%s)'%(sum(predict_train),sum(solutions),len(dataset)))
        print("Correct positives: %s\nCorrect negatives: %s\nFalse positives: %s\nFalse negatives: %s" % (tp, tn, fp, fn))

    # Accuracy Score on train dataset
    if acc_met == "f1":
        try:
            p=tp/(tp+fp)
            r=tp/(tp+fn)
            accuracy_train=2*p*r/(p+r)
        except ZeroDivisionError:
            accuracy_train = 0
    else:
        accuracy_train = accuracy_score(solutions,predict_train)
    if do_print:
        print('accuracy_score on dataset : %s\n' % accuracy_train)
    if get_acc:
        return accuracy_train
    return predict_train
    
def train_superforest(inputs, solutions, SUPERFOREST_SIZE=25, jobs=6, est=20):
    return Parallel(n_jobs=jobs)(delayed(pforest)(inputs, solutions, est=est) for _ in tqdm_notebook(range(SUPERFOREST_SIZE)))

def train_superforest2(inputs, solutions, SUPERFOREST_SIZE=25, verbose=0):
    models = list()
    for i in range(SUPERFOREST_SIZE):
        models.append(pmethod(inputs, solutions, verbose=verbose))
    return models
```

```python
import numpy as np

def filter_model(n, inputs, solutions, cutoff=0.5):
    print("stack height: %s" % n)
    if n!=0:
        superforest = pmethod(inputs, solutions, verbose=0)
    else:
        superforest = train_superforest(inputs, solutions, SUPERFOREST_SIZE=1)
    predictions = get_superforest_predictions(superforest, inputs, solutions=solutions, cutoff=cutoff)
    if n == 0:
        out = list()
        out.append(superforest)
        return out
    assert len(predictions) == len(inputs) == len(solutions)
    for i in range(len(solutions)-1, -1, -1):
        if rn(predictions[i], cutoff=cutoff) == 1:
            solutions = np.delete(solutions, i, 0)
            inputs = np.delete(inputs, i, 0)
    out = filter_model(n-1, inputs, solutions, cutoff=cutoff)
    out.append(superforest)
    return out

l_sf = filter_model(1, X_train, y_train)
```
