```python
import pandas as pd

df_training_full = pd.read_csv("training.csv", na_values=['na'])

df_training_full.head()
```

```python
df_test = pd.read_csv("test.csv", na_values=['na'])
df_test.head()
```

```python
#Combined both Train and Test Data set to do preprocessing together # and set flag for both as well
df_training_full['Type'] = 'Train' 
df_test['Type'] = 'Test'
fullData = pd.concat([df_training_full,df_test],axis=0,sort=True)
fullData.dtypes
```

```python
fullData.head()
```

```python
numeric_cols = list()
for k,v in fullData.iteritems():
    if v.dtype in ['int64', 'float64']:
        numeric_cols.append(k)


fullData[numeric_cols] = fullData[numeric_cols].fillna(fullData[numeric_cols].mean())
fullData.head()
```

```python
print("numeric columns:", numeric_cols)
# this spams
```

```python
# normalize
fullData[numeric_cols] = fullData[numeric_cols]/fullData[numeric_cols].max()
fullData.head()
```

```python
from sklearn.model_selection import train_test_split

train=fullData[fullData['Type']=="Train"]
test=fullData[fullData['Type']=="Test"]

numeric_cols.remove("target") #no answers kek
X = train[numeric_cols].values
y = train["target"].values

X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
#import necessary libraries to build model
import random
from sklearn.ensemble import RandomForestRegressor
from tqdm import tqdm
from joblib import Parallel, delayed

random.seed(42)
rf_list = list()
SUPERFOREST_SIZE = 25  # n x microforests

def pmethod():
    x_subset, _, y_subset, _ = train_test_split(X_train, y_train, test_size=0.7, random_state=42)
    rf = RandomForestRegressor(n_estimators=10)
    rf.fit(x_subset, y_subset)
    return rf
    
rf_list = Parallel(n_jobs=6)(delayed(pmethod)() for _ in tqdm(range(SUPERFOREST_SIZE)))
```

```python
from sklearn.metrics import accuracy_score

def rn(n):
    if n > 0.5:
        return 1
    else:
        return 0

def calc_stats(Vx, Vy):
    """
    Vx: predictions
    Vy: actual
    """
    assert len(Vx) == len(Vy), "length mismatch: %s/%s" % (len(Vx), len(Vy))
    cp=0;cn=0;fp=0;fn=0
    for i in range(len(Vx)):
        if rn(Vx[i]) == 1 and rn(Vy[i]) == 1:
            cp+=1
        elif rn(Vx[i]) == 0 and rn(Vy[i]) == 0:
            cn+=1
        elif rn(Vx[i]) == 1 and rn(Vy[i]) == 0:
            fp+=1
        elif rn(Vx[i]) == 0 and rn(Vy[i]) == 1:
            fn+=1
        else:
            raise Exception("wtf")
    return cp, cn, fp, fn

def coag_superforest(sf, data):
    # print("data length: ", len(data))
    predict_train_list = list()
    for rf in rf_list:
        predict_train_list.append(rf.predict(data))

    predict_train = list()
    for i in range(len(data)):
        s=0
        for pt in predict_train_list:
            s+=pt[i]
        predict_train.append(rn(s/len(predict_train_list)))
    return predict_train

# number of trees used
print('Number of Trees used : ', rf.n_estimators)

# predict the target on the train dataset
predict_train = coag_superforest(rf_list, X_train)
print('Targets found on train data %.2f/%.2f'%(sum(predict_train),sum(y_train)))
print("Correct positives: %s\nCorrect negatives: %s\nFalse positives: %s\nFalse negatives: %s" % calc_stats(predict_train, y_train))

# Accuray Score on train dataset
accuracy_train = accuracy_score(y_train,predict_train)
print('accuracy_score on train dataset : %s\n' % accuracy_train)

# predict the target on the test dataset
predict_test = coag_superforest(rf_list, X_valid)
print('Targets found on test data %.2f/%.2f'%(sum(predict_test),sum(y_valid))) 
print("Correct positives: %s\nCorrect negatives: %s\nFalse positives: %s\nFalse negatives: %s" % calc_stats(predict_test, y_valid))

# Accuracy Score on test dataset
accuracy_test = accuracy_score(y_valid,predict_test)
print('accuracy_score on test dataset : ', accuracy_test)
```
